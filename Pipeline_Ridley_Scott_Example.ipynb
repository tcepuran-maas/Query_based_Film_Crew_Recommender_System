```python
# Example: Ridley Scott movies – find the best Actor + Composer combination
filtered_crew = queryPrefiltering(
    df=crew_df,
    movies_df=movies_df,
    req_person_names=["Ridley Scott"],
    req_job_titles=["Director"],
    needed_jobs=["Actor", "Original Music Composer"],
    genres=None,   # e.g. ["Science Fiction"] for genre filtering
    actor_cast_max=500,     # only Actor with cast_id < 20
    max_candidates_per_job={
        "Actor": 25,
        "Original Music Composer": 20
    }
)
filtered_crew
```

```python
from Build_Graph import generateGraph
g = generateGraph(output_path='data/test.ttl',movie_crew_df=filtered_crew)
print(len(g))
```

```python
sparql_query = '''
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ex: <http://example.org/>
PREFIX schema: <https://schema.org/>
PREFIX TMDB: <https://www.themoviedb.org/>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?movieName ?ratingValue
WHERE {
    ?movie a schema:Movie .
    ?movie schema:name ?movieName .
    ?movie schema:aggregateRating ?rating .
    ?rating schema:ratingValue ?ratingValue .

    ?movie schema:genre ?genre .
    ?genre schema:name ?genreName .
    FILTER(LCASE(STR(?genreName)) = "science fiction")

    ?pDir ex:inMovie ?movie ;
          ex:hasParticipant ?director ;
          ex:hasRole ?rDir .
    ?director schema:name "Ridley Scott"^^xsd:string .
    ?rDir schema:roleName "Director"^^xsd:string .
}
ORDER BY DESC(?ratingValue)
LIMIT 10
'''

results = g.query(sparql_query)

# print results
rows = [tuple(r) for r in results]
print(f'Matches: {len(rows)}')
for row in rows[:25]:
    print(row)
```

```python

sparql_query = '''PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ex: <http://example.org/>
PREFIX schema: <https://schema.org/>
PREFIX TMDB: <https://www.themoviedb.org/>

SELECT ?directorName ?actressName (COUNT(DISTINCT ?movie) AS ?score) (AVG(?ratingValue) AS ?averageRating)
WHERE {

  {
    SELECT ?movie ?director
    WHERE {
      ?movie a schema:Movie ;
             schema:genre ?genre .


      ?pDir ex:inMovie ?movie ;
            ex:hasParticipant ?director ;
            ex:hasRole ?rDir .
      ?director schema:name "Ridley Scott"^^xsd:string .
      ?rDir schema:roleName "Director"^^xsd:string .
    }
  }

  # Filter actresses (Lead Actor)
  ?pActress ex:inMovie ?movie ;
            ex:hasParticipant ?actress ;
            ex:hasRole ?rActress ;
            ex:castOrder ?order .
  ?rActress schema:roleName "Actor"^^xsd:string .
  ?actress schema:gender "f"^^xsd:string ;
           schema:name ?actressName .
    ?movie schema:aggregateRating ?rating .
    ?rating schema:ratingValue ?ratingValue .


  # Director name
  ?director schema:name ?directorName .
}
GROUP BY ?directorName ?actressName
ORDER BY DESC(?averageRating) DESC(?score) 
LIMIT 3
'''

results = g.query(sparql_query)

# print results
rows = [tuple(r) for r in results]
print(f'Matches: {len(rows)}')
for row in rows[:25]:
    print(row)
```

```python
import time

sparql_query = '''
PREFIX ex:     <http://example.org/>
PREFIX schema: <https://schema.org/>
PREFIX xsd:    <http://www.w3.org/2001/XMLSchema#>

SELECT ?actorName ?composerName ?ridleyActorMovies ?ridleyComposerMovies ?crossCollabMovies
       (?ridleyActorMovies + ?ridleyComposerMovies + ?crossCollabMovies AS ?totalScore)
WHERE {
  # Actors who worked with Ridley Scott as Director
  {
    SELECT ?actor (COUNT(DISTINCT ?mActor) AS ?ridleyActorMovies)
    WHERE {
      ?pDir ex:inMovie ?mActor ; ex:hasParticipant ?ridley ; ex:hasRole ?rDir .
      ?ridley schema:name "Ridley Scott"^^xsd:string .
      ?rDir schema:roleName "Director"^^xsd:string .

      ?pAct ex:inMovie ?mActor ; ex:hasParticipant ?actor ; ex:hasRole ?rAct .
      ?rAct schema:roleName "Actor"^^xsd:string .
      # optional (if needed globally):
      # ?actor schema:gender "f"^^xsd:string .
    }
    GROUP BY ?actor
  }

  # Composers who worked with Ridley Scott as Director
  {
    SELECT ?composer (COUNT(DISTINCT ?mComposer) AS ?ridleyComposerMovies)
    WHERE {
      ?pDir ex:inMovie ?mComposer ; ex:hasParticipant ?ridley ; ex:hasRole ?rDir .
      ?ridley schema:name "Ridley Scott"^^xsd:string .
      ?rDir schema:roleName "Director"^^xsd:string .

      ?pComp ex:inMovie ?mComposer ; ex:hasParticipant ?composer ; ex:hasRole ?rComp .
      ?rComp schema:roleName "Original Music Composer"^^xsd:string .
    }
    GROUP BY ?composer
  }

  # Cross-collabs between those actor/composer candidates
  {
    SELECT ?actor ?composer (COUNT(DISTINCT ?mCross) AS ?crossCollabMovies)
    WHERE {
      ?pAct ex:inMovie ?mCross ; ex:hasParticipant ?actor ; ex:hasRole ?rAct .
      ?rAct schema:roleName "Actor"^^xsd:string .
      ?actor schema:gender "f"^^xsd:string .

      ?pComp ex:inMovie ?mCross ; ex:hasParticipant ?composer ; ex:hasRole ?rComp .
      ?rComp schema:roleName "Original Music Composer"^^xsd:string .
    }
    GROUP BY ?actor ?composer
  }

  ?actor schema:name ?actorName .
  ?composer schema:name ?composerName .
}
ORDER BY DESC(?totalScore)
LIMIT 20
'''

start = time.perf_counter()
results = g.query(sparql_query)
elapsed = time.perf_counter() - start

rows = [tuple(r) for r in results]
print(f"Matches: {len(rows)} | Runtime: {elapsed:.3f}s")
print(f"{'Actor':<30} {'Composer':<35} {'w/ Ridley (A)':<15} {'w/ Ridley (C)':<15} {'A↔C Cross':<12} {'Total':>7}")
print("-" * 115)
for row in rows:
    print(f"{str(row[0]):<30} {str(row[1]):<35} {str(row[2]):<15} {str(row[3]):<15} {str(row[4]):<12} {str(row[5]):>7}")
```
