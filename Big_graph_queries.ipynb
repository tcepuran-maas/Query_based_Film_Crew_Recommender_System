```python
from pathlib import Path
import rdflib

# project path relative to the notebook
ttl_path = Path('data') / 'film_crew.ttl'

g = rdflib.Graph()
g.parse(ttl_path, format='turtle')
print(f'Graph loaded: {len(g)} triples from {ttl_path}')
```

```python
q = '''
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>

SELECT ?class (COUNT(DISTINCT ?instance) AS ?count)
WHERE {
  ?instance a ?class .
  ?class a owl:Class .
}
GROUP BY ?class
ORDER BY DESC(?count)
'''

rows = list(g.query(q))
print('rows', len(rows))
for r in rows:
    print(r[0], r[1])
```

```python
sparql_query = '''
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX ex: <http://example.org/>
PREFIX schema: <https://schema.org/>
PREFIX TMDB: <https://www.themoviedb.org/>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>

SELECT ?movieName ?ratingValue
WHERE {
    ?movie a schema:Movie .
    ?movie schema:name ?movieName .
    ?movie schema:aggregateRating ?rating .
    ?rating schema:ratingValue ?ratingValue .

    ?movie schema:genre ?genre .
    ?genre schema:name ?genreName .
    FILTER(LCASE(STR(?genreName)) = "science fiction")

    ?pDir ex:inMovie ?movie ;
          ex:hasParticipant ?director ;
          ex:hasRole ?rDir .
    ?director schema:name "Ridley Scott"^^xsd:string .
    ?rDir schema:roleName "Director"^^xsd:string .
}
ORDER BY DESC(?ratingValue)
LIMIT 10
'''

results = g.query(sparql_query)

# print results
rows = [tuple(r) for r in results]
print(f'Matches: {len(rows)}')
for row in rows[:25]:
    print(row)
```
